# Exercise 1: Read Total profit of all months and show it using a line plot
Total profit data provided for each month. Generated line plot must include the following properties:
- X label name = Month Number
- Y label name = Total profit

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('company_sales_data.csv')
```

```python
df.head()
```

```python
x = df['month_number']
y = df['total_profit']
plt.plot(x,y)
plt.xlabel('Month Number')
plt.ylabel('Total Profit')
plt.xlim(1,12)
plt.suptitle('Total Profit vs Months')
```

# Exercise 2: Get total profit of all months and show line plot with the following Style properties
Generated line plot must include following Style properties:
- [x] Line Style dotted and Line-color should be red
- [x] Show legend at the lower right location
- [x] X label name = Month Number
- [x] Y label name = Sold units number
- [x] Add a circle marker
- [x] Line marker color as red
- [x] Line width should be 3
- References: [matplotlib.lines.Line2D](https://matplotlib.org/stable/api/_as_gen/matplotlib.lines.Line2D.html)

```python
x = df['month_number']
y = df['total_profit']
plt.plot(x,y,'r:',linewidth=3.0,marker='o',label='Sold units number')
plt.xlabel('Month Number')
plt.ylabel('Total Profit')
plt.legend(loc='lower right')
plt.xlim(1,12)
plt.suptitle('Number of units sold per month')
```

# Exercise 3: Read all product sales data and show it using a multiline plot
Display the number of units sold per month for each product using multiline plots. (i.e., Separate Plotline for each product).

# Exercise 4: Read toothpaste sales data of each month and show it using a scatter plot
- Also, add a grid in the plot. gridline style should .

```python
x_month = df['month_number']
y_products = df['toothpaste']
plt.scatter(x_month,y_products,)
plt.grid(linestyle='-')
plt.suptitle('Toothpaste Sold per Month')
plt.xlabel('Month')
plt.ylabel('Total Units Sold')
```

# Exercise 5: Read face cream and facewash product sales data and show it using the bar chart

```python
x = df[['facecream','facewash']]
# x.columns.to_list()
plt.bar(x.columns.to_list(),height=x.sum())
plt.grid(linestyle='-')
plt.suptitle('Total unit sold for facecream and facewash')
plt.xlabel('Product Name')
plt.ylabel('Total unit sold')
```

# Exercise 9: Read Bathing soap facewash of all months and display it using the Subplot

```python
bathing_soap = df['bathingsoap']
face_wash = df['facewash']

# subplot for bathing soap 
fig, ax = plt.subplots(2,1)
ax[0].plot(df['month_number'],bathing_soap)
ax[0].set_title('bathing soap')
ax[0].set_xlabel('month')
ax[0].set_ylabel('total unit sold')

# subplot for facewash
ax[1].plot(df['month_number'],face_wash)
ax[1].set_title('face wash')
ax[1].set_xlabel('month')
ax[1].set_ylabel('total unit sold')

#figure
fig.tight_layout(pad=1)
```

# Exercise Question 10: Read all product sales data and show it using the stack plot
- Use transpose: https://pandas.pydata.org/docs/reference/api/pandas.DataFrame.transpose.html

```python
# transpose the data so that the dimensions are the same
y = df[['facecream','facewash','toothpaste','bathingsoap','shampoo','moisturizer']].transpose()
x = df['month_number']

# plot
#fig, ax = plt.subplots()
#ax.stackplot(x,y)
plt.stackplot(x,y,labels=['facecream','facewash','toothpaste','bathingsoap','shampoo','moisturizer'])
plt.legend()
plt.xlim(1,12)
plt.suptitle('Stackplot of product sold per month')
plt.xlabel('months')
plt.ylabel('Total unit sold')
```
